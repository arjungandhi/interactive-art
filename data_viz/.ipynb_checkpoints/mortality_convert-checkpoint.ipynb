```python
import re
import json
```

```python
mortality_text = open('mortality.txt','r').read()
```

```python
d = mortality_text.split('\n')
del(d[0])
```

```python
final_dict = {}
for entry in d:
    entry=entry.split('\t')
    if len(entry) == 8:
        for i,v in enumerate(entry):
            entry[i] = v.replace('"','')
        if entry[1] not in final_dict.keys():
            final_dict[entry[1]] = {}
        if 'NS' not in entry[4]:
            key = re.split(r"[^0-9]", entry[4])
            key = [a for a in key if a != '']
            if key[-1] == '85':
                key[-1] == '100'
            final_dict[entry[1]][key[-1]] = float(entry[-1])/100000
    
```
